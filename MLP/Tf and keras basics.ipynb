```
import tensorflow as tf
from tensorflow import keras
```

```
model = tf.keras.Sequential()
model.add(10,input_shape=(784,),kernel_initializer='zeros',name='dense_layer',activation='softmax')    #Creates a model with 10 nodes, taking in an input of shape (784,) , using a kernel_initializer of 'zeros' (will assign all the weights of nodes as zeros. can use 'random_uniform' to randomly assign the weights to -0.05 to 0.05)
```
