```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorboard
print(tf.__version__)
```

```
(x_train,y_train) , (x_eval,y_eval) = tf.keras.datasets.mnist.load_data()
```

```
#Normalize feature data and convert to float32 type

x_train = x_train / np.float32(255)
y_train = y_train.astype(np.int32)

x_eval = x_eval / np.float32(255)
y_eval = y_eval.astype(np.int32)
```

```
feature_columns = [tf.feature_column.numeric_column('x',shape=[28,28])]
```

```
#Create the estimator, feeding the feature_cols , no. of classes to predict and where to save the parameteres to

classifier = tf.estimator.LinearClassifier(feature_columns = feature_columns, n_classes=10,model_dir='mnist_model/')
```

```
# Create input function to feed into the estimator during training (to feed the input values to the estimators). The input function generates tensors that can be fed to estimator

train_input_function = tf.compat.v1.estimator.inputs.numpy_input_fn(x={'x':x_train}, y=y_train ,batch_size=100 ,num_epochs=None ,shuffle=True)
```

```
classifier.train(train_input_function, steps=10)
```

```
# Create input function for eval training set. (tf.compat.v1.estimator.inputs.numpy_input_fn will return input function that would feed dict numpy arrays into the model)

eval_input_function = tf.compat.v1.estimator.inputs.numpy_input_fn(x={'x':x_eval}, y=y_eval ,num_epochs=1 ,shuffle=False)
```

```
eval_results = classifier.evaluate(eval_input_function)
print(eval_results)
```
