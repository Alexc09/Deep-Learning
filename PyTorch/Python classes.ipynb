```
class rect:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def get_area(self):
    return (self.length * self.width)
```

```
r = rect(3,4)
r.get_area()
```

```
class square(rect):
  def __init__(self, length):
    #Inherit from the rect class
    super().__init__(length, length)
```

```
s = square(3)
s.get_area()
```

```
def decorator_fn(original_fn):
  def wrapper_fn():
    return original_fn()
  return wrapper_fn

def disp():
  print('Ran the disp() function!')

wrapped = decorator_fn(disp)

wrapped()
```

```
def decoratorf(original):
  def wrapper():
    print('decorator function running')
    return original()
    
  return wrapper
```

```
@decoratorf
#This basically means you set disp() = decoratorf(disp)
#You will set the disp() function to be the function where disp() is wrapped by decoratorf()
def disp():
  print('this is the disp function running')

#Same as
# disp = decorator(disp)
```

```
disp()
```

```
#Must add *args and **kwargs to enable function to accept multiple input arguments
def decoratorf(original):
  def wrapper(*args, **kwargs):
    print(f'decorator function running {original.__name__}')
    return original(*args, **kwargs)

  return wrapper
```

```
@decoratorf
def show(name):
  print(f'Show function ran with name {name}')
```

```
show('Tet')
```

```
class decorator_class:
  def __init__(self, original_fn):
    self.original_fn = original_fn

  def __call__(self, *args, **kwargs):
    print(f'decorator_class ran while executing {self.original_fn.__name__} function')
    return self.original_fn(*args, **kwargs)
```

```
@decorator_class
def show_this(name):
  print(f'THis function was ran by {name}')

show_this('jake')
```

```
decorator_class(show_this('jake'))
```

```
import time

def time_this(original_fn):
  def wrapper(*args, **kwargs):
    t = time.time()
    result = original_fn(*args, **kwargs)
    t1 = time.time() - t
    print(f'Time to run {original_fn.__name__} is {t1}')
  
  return wrapper
```

```
@time_this
def sef(name):
  print(f'How long will it take to run {name}')
```

```
sef('Jake')
```

```
class Student:
  def __init__(self, first, last):
    self.first = first
    self.last = last
    self.mail = 'gmail'
  
  #With the @property I'm able to access the email function as an attribute
  #This is a getter
  @property
  def email(self):
    return f'{self.first}.{self.last}@{self.mail}.com'

  #Setter 
  @email.setter
  def email(self, mail):
    self.mail = mail

  #Deleter
  @email.deleter
  def email(self):
    print(f'Deleted the mail {self.mail}!')
    self.mail = None
```

```
x = Student('Jake', 'rou')
x.email
```

```
#Without the @email.setter, you won't be able to do this (directly set the mail using the email method)
x.email = 'hotmail'
```
